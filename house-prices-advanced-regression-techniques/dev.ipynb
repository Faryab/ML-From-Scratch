```python
from sklearn.model_selection import KFold, cross_val_score
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import OneHotEncoder
from sklearn.impute import SimpleImputer
from sklearn.linear_model import LinearRegression
```

```python
# Load data
target = "SalePrice"
train = pd.read_csv("data/train.csv")
test = pd.read_csv("data/test.csv")  # keep for later submission

# Split features/target; model log(price) because the competition scores RMSE(log(price))
X = train.drop(columns=[target])
y = np.log1p(train[target])
```

```python
# Separate numeric vs categorical columns so we can apply different preprocessing
num_cols = X.select_dtypes(include=["int64", "float64"]).columns
cat_cols = X.select_dtypes(include=["object"]).columns
```

```python
# Numeric preprocessing:
# - Median imputation handles missing numeric values robustly (less sensitive to outliers than mean)
num_pipe = Pipeline(
    steps=[
        ("imputer", SimpleImputer(strategy="median"))
    ]
)


# Categorical preprocessing:
# - Fill missing categories with a literal "None" meaning "feature absent/unknown"
# - One-hot encode turns categories into 0/1 columns; ignore unseen categories at test time
cat_pipe = Pipeline(steps=[
    ("imputer", SimpleImputer(strategy="constant", fill_value="None")),
    ("onehot", OneHotEncoder(handle_unknown="ignore"))
])

# Combine the two pipelines column-wise:
# - Applies numeric pipeline to numeric columns
# - Applies categorical pipeline to categorical columns
preprocess = ColumnTransformer(transformers=[
    ("num", num_pipe, num_cols),
    ("cat", cat_pipe, cat_cols)
])

# Simplest reasonable model:
# - Ordinary least squares linear regression (no regularization)
model = LinearRegression()

# Full pipeline ensures preprocessing is fit only on training folds during CV (prevents leakage)
pipe = Pipeline(steps=[
    ("preprocess", preprocess),
    ("model", model)
])

# Cross-validation setup:
# - Shuffle for stability; fixed seed for reproducibility
cv = KFold(n_splits=5, shuffle=True, random_state=42)

# Kaggle metric is RMSE on log-scale; sklearn returns negative for "neg_*" scorers
rmse = -cross_val_score(estimator=pipe, X=X, y=y, scoring="neg_root_mean_squared_error", cv=cv)
rmse.mean(), rmse.std()
```

```python
pipe.fit(X, y)

test_pred_log = pipe.predict(test)
test_pred = np.expm1(test_pred_log)          # back to dollars
test_pred = np.clip(test_pred, 0, None)      # avoid negative predictions (can happen with linear models)

submission = pd.DataFrame({
    "Id": test_ids,
    "SalePrice": test_pred,
})

submission.to_csv("submission.csv", index=False)
submission.head()
```
